Fordi K'(t)=k'K(t) der k=0.05 gir at veksten per m√•ned er fem prosent av befolkningen. Initialbetingelsen er K(0)=100

```python
from pylab import *

h=1E-2
N=int(12/h)
y=zeros(N)
k=0.05
x=zeros(N)
y[0]=200
def yder(y,x):
    return k*y

for i in range(N-1):
    y[i+1]=y[i]+yder(y[i],x[i])*h
    x[i+1]=x[i]+h

plot(x,y)
grid('True')
```

```python
from pylab import *

h=1E-2
N=int(36/h)
y=zeros(N)
k=0.18
x=zeros(N)
y[0]=5000
b=50000
def yder(y,x):
    return k*y*(1-y/b)

for i in range(N-1):
    y[i+1]=y[i]+yder(y[i],x[i])*h
    x[i+1]=x[i]+h

plot(x,y)
grid('True')
```
